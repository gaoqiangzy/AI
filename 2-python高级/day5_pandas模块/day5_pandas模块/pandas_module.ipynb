```python
# apply1=pd.read_csv("./files/apply1.csv",index_col=0)
# def f5(x):
#     return pd.Series(x.split('-'))
# #     return 0
# apply1['E'].apply(f5)
```

```python
# apply2=pd.read_csv("./files/apply2.csv")
# apply2.groupby('address')[['age','money']].mean()
# apply2.groupby("age")['money'].mean()
# apply2.groupby('sex')['money'].mean()
# apply2.groupby(['address','sex'])['money'].mean().unstack(level=1)
# apply2.groupby('address')['money'].agg(['max','min','mean'])

def f6(x):
    rmax=x.max()
    rmin=x.min()
    rmean=x.mean()
    rcha=rmax-rmin
    return pd.Series([rmax,rmin,rmean,rcha],index=['最大值','最小值','平均值','差'])
apply2.groupby('address')['money'].apply(f6).unstack(level=1)
```

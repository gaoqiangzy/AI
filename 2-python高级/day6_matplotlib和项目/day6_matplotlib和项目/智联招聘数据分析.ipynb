```python
def f2(x):
    try:
        res=re.search("((?:\.|\d)+)K-((?:\.|\d)+)K",x).groups()
        result=(float(res[0])+float(res[1]))/2*1000
        if result>=0 and result<8000:
            return "1-8k"
        elif result>=8000 and result<15000:
            return "8-15k"
        elif result>=15000 and result<22000:
            return "15-22k"
        elif result>=22000 and result<30000:
            return "22-30k"
        elif result>=30000 and result<50000:
            return "30-50k"
        else:
            return "50K以上"
    except :
        return "1-8K"
# zhilian['salary_num']=zhilian['salary'].apply(f2)
```

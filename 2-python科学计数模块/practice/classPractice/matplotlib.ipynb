```python
import matplotlib.pyplot as plt
# 挂掉   tk在图形化gui界面中绘制
# %matplotlib tk  
# %matplotlib inline      在浏览器中绘制
import numpy as np
```

```python
x=[1,2,3,4,5]
y=np.random.randint(0,10,5)
#绘图 plot() 折线图
plt.plot(x,y)
plt.show()
```

## 颜色。标记，线形

```python
# figure（"图形名",figsize=(长,宽)）图形 画板  针对tk
#plt.plot(x,y,color,linestyle="",linewidth= ,marker="",markersize= ,lable="")
#linestyle="-" 实线 "--"虚线 "-."点杠线 ":"点线
# linewidth  线宽度   marker 标记样式  markersize 样式大小
plt.figure('3')
x1=np.arange(-3,3,0.2)
y1=np.cos(x1)
plt.plot(x,y,color="black",linestyle="--",linewidth=5,label="public")
plt.plot(x1,y1,color="red",linestyle=":",marker="*",markersize=10,label="cos")
#生成图例 legend()
plt.legend()
plt.show()
```

## 刻度，标题，标签，图例

```python
# 以下三行解决中文乱码问题
import matplotlib as mpl
mpl.rcParams['font.sans-serif'] = ['SimHei']
mpl.rcParams['axes.unicode_minus'] = False

x=np.arange(8)
Chinese=np.random.randint(60,101,8)
Math=np.random.randint(60,101,8)
plt.plot(x,Chinese,label="语文成绩")
plt.plot(x,Math,label="数学成绩")
# plt.xlim()  plt.ylin()  限制xy的值
plt.ylim(0,100)
ax=plt.gca() #获取当前的坐标轴
ax.spines['right'].set_color("none")  #去除右边边框
ax.spines['top'].set_color("none")
plt.ylabel("成绩（分）")
plt.xlabel("阶段(学期)")
plt.title("小明大学各个学期的分数")

#xticks()  yticks() 设置某个值上的刻度值
x_data=['大一上','大一上','大二上','大二下','大三上','大三下','大四上','大四下']
plt.xticks(x,x_data)  #将x轴的各左表名用x_data内数据表示
#plt.xlabel()  plt.ylabel() 设置xy轴标题
plt.legend()
plt.show()
```

## 子图 subplot

```python
English=np.random.randint(60,101,8)
# 第一个子图 两行两列第一个
plt.subplot(221)
plt.plot(x,Chinese)
plt.subplot(222)
plt.plot(x,Math)
plt.subplot(223)
plt.plot(x,English)
plt.show()
```

## 常见图形

```python
# 柱状图
plt.bar(x-0.2,Chinese,width=0.4,label="语文成绩")
plt.bar(x+0.2,Math,width=0.4,label="数学成绩")
plt.xticks(x,x_data)
plt.legend()
#给树状图添加数值显示 ，使用text方法
for i in range(len(x)):
    plt.text(x[i]-0.4,Chinese[i]+1,Chinese[i])
    plt.text(x[i]+0.1,Math[i]+1,Math[i])
plt.show()
```

```python
#散点图 sactter
import pandas as pd
man=pd.read_csv("../../day5_pandas模块/day5_pandas模块/files/user_info_man.csv",sep="::",header=None,encoding="utf-8",engine="python")
woman=pd.read_csv("../../day5_pandas模块/day5_pandas模块/files/user_info_woman.csv",sep="::",header=None,encoding="utf-8",engine="python")
# man
# woman
plt.scatter(man[10],man[11])
plt.scatter(woman[10],woman[11])
plt.xlabel("身高")  #x轴名称
plt.ylabel("体重")   #y轴名称
plt.xlim(100,250)   #x轴范围
plt.ylim(0,300)    #y轴范围
plt.show()
```

```python
#饼图 pie
#plt.pie(sata,labels="",explode="",startangle= ,shadow=True,labeldistance= , autopct="")
# labels 每一块的文字显示
# explode 每一块离开中心的距离，默认0
#startangle  绘图开始的角度，默认0
#shadow  阴影    labeldistance  label放置的位置 >1  放在图形外面  默认1.1
#autopct  百分比  pctdistance 百分比放置的位置   radius  图形半径  
x=man[8].value_counts().values
index=man[8].value_counts().index
explode=np.zeros(12)
explode[x==x.max()]=0.1
plt.pie(x,labels=index,explode=explode,startangle=60,shadow=True,labeldistance=1.1,autopct="%.1f%%",pctdistance=0.7,radius=1.8)
plt.show()
```

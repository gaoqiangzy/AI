## 排序

```python
# sort_index()索引排序
scores=pd.DataFrame(np.random.randint(0,101,[5,4]),index=list("bcdae"),columns=list("wzyx"))
scores.sort_index() #对行索引排序 默认升序
scores.sort_index(ascending=False) #对行索引降序
scores.sort_index(axis=1,ascending=False) #对列进行降序

#sort_values() 值排序
scores.sort_values(by='w') #对w列进行升序
scores.sort_values(by=['w','z'],ascending=[False,True]) #对w列降序排列 碰到相同值 再按照z列升序排列

#rank() 排名
scores.rank()
scores.rank(method="min") #共用最小排名
scores.rank(method="first") #不出现相同排名 排列靠前的排名更高
```

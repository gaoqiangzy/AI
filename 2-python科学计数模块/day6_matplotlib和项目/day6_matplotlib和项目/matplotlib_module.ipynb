```python
import matplotlib.pyplot as plt
# tk 在图形化gui界面中绘制图形
# inline 在浏览器中绘图
# %matplotlib tk
import numpy as np
import matplotlib as mpl
mpl.rcParams['font.sans-serif'] = ['SimHei']
mpl.rcParams['axes.unicode_minus'] = False
import pandas as pd
```

```python
# 数据
x=[1,2,3,4,5]
y=np.random.randint(0,10,5)
# 绘图 plot() 折线图
plt.plot(x,y)
# 显示图形  show() 
plt.show()
```

```python
# figure("图形名",figsize=(长,宽)) 图形,画板 大小单位为英寸
# plt.plot(x,y,color,linestyle="")
# linestyle="-" 线条的样式 - 实线  -- 虚线  -. 点线
# linewidth=num  默认1
# marker=""  数据点的样式
# label='' 设置线条的标签

plt.figure('第3个画板',figsize=(8,6))
y=np.random.randint(0,10,5)
x1=np.arange(-3,3,0.2)
y1=np.cos(x1)
# plt.plot(x,y,x1,y1)
plt.plot(x,y,color="black",linestyle='-.',linewidth=1,marker="o",markersize=5,label="public")
plt.plot(x1,y1,color="red",linestyle=':',marker="*",markersize=5,label="cos")
# 生成图例 legend()
plt.legend()


plt.show()
# plt.figure("第四个画板",figsize=(8,6))
# plt.plot(x,y)
# plt.show()
```

```python
plt.figure()
x=np.arange(8)
Chinese=np.random.randint(60,101,8)
Math=np.random.randint(50,101,8)
plt.plot(x,Chinese,label="语文成绩")
plt.plot(x,Math,label="数学成绩")
# plt.xlim()  plt.ylim() 设置范围
plt.ylim(0,100)
ax=plt.gca() # 获取当前的坐标轴
ax.spines['right'].set_color("none")
ax.spines['top'].set_color("none")
# plt.xlabel() plt.ylabel( ) 设置x,y轴标题
plt.ylabel("成绩(分)")
plt.xlabel("阶段(学期)")
plt.title("小明的各个学期的分数")

# xticks()  yticks()  设置在某个值上的刻度值
x_data=['大一上','大一下','大二上','大二下','大三上','大三下','大四上','大四下']
plt.xticks(x,x_data)


plt.legend()
plt.show()
```

```python
# 散点图 scatter
man=pd.read_csv("./files/user_info_man.csv",sep="::",encoding='utf-8',engine='python',header=None)
man
woman=pd.read_csv("./files/user_info_woman.csv",sep="::",encoding='utf-8',engine='python',header=None)
woman

plt.figure("身高体重散点图",figsize=(8,6))
plt.scatter(man[10],man[11],alpha=1,marker="*")
plt.scatter(woman[10],woman[11],alpha=1)
plt.xlim(100,250)
plt.ylim(0,200)
plt.xlabel("身高")
plt.ylabel("体重")
plt.show()
```

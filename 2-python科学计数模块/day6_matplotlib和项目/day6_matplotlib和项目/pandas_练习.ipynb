```python
# 案例三
user=pd.read_csv("../../day6_matplotlib和项目/day6_matplotlib和项目/files/案例三/ml-1m/users.dat",sep="::",engine="python",names=['uid','sex','age','job','zipCode'])
user
movie=pd.read_csv("../../day6_matplotlib和项目/day6_matplotlib和项目/files/案例三/ml-1m/movies.dat",sep="::",engine="python",names=['mid','title','Genres'])
movie
rating=pd.read_csv("../../day6_matplotlib和项目/day6_matplotlib和项目/files/案例三/ml-1m/ratings.dat",sep="::",engine="python",names=['uid','mid','rating','timestamp'])
rating
rating['uid']=rating['uid'].astype('uint16')
rating['mid']=rating['mid'].astype('uint16')
rating['rating']=rating['rating'].astype('uint8')
rating['timestamp']=rating['timestamp'].astype('uint32')
rating.info()
rating
```

```python
# 查看不同性别对电影的评分
user_rating=pd.merge(user,rating,left_on="uid",right_on='uid')
user_rating.groupby(['mid','sex'])['rating'].mean()
```

```python
# 不同性别差异最大的电影  var()求方差
mid=user_rating.groupby(['mid','sex'])['rating'].mean().var(level=0).sort_values(ascending=False).head(1).index.values[0]
movie[movie['mid']==mid]
```

```python
# 查找出热门电影  热度榜单
movie_rating=pd.merge(movie,rating,left_on='mid',right_on='mid')
movie_rating
# rating['mid'].value_counts()
hot_rank=movie_rating.groupby('title')['title'].count().sort_values(ascending=False)
```

```python
# 每部电影的评分（平均分）  高分榜单
rating_rank=movie_rating.groupby('title')['rating'].mean().sort_values(ascending=False)
```

```python
# 前十大热门电影的评分
rating_rank.loc[hot_rank.head(10).index.values]
```
